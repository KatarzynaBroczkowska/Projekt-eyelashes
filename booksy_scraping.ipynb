# Scraping

```python
from bs4 import BeautifulSoup
import requests
```

```python
from tqdm import tqdm

# total number of pages
page_count = 209
# query from url
query = 'laminacja%20rzes'

tmp = []
pages = range(1, page_count + 1)

for page in tqdm(pages):
    url = f'https://booksy.com/pl-pl/s?businessesPage={page}&query={query}'
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html')
    search_data = soup.find_all('ul')[1]
    items = search_data.find_all('li')

    for item in items:
        # location
        loc = item.find_all('a')[1].find_all('div')[-1].text.strip()

        names = item.find_all('div', class_='purify_aIF50SidLa4hohq9cO8NHw==')
        durations = item.find_all('div', class_='purify_lPTOnTJKdGtFsWiIA3TUBA==')
        prices = item.find_all('div', class_='purify_s8fylQf1Z9K4N9kwPLHJaA== purify_IbX1Bb-sl2ffhqyi9xUH1A== purify_f9tJlQGTtqZe8Hg0f7oQhg==')
        for nm, dur, pri in zip(names, durations, prices):
            name = nm.find('h3').text.strip()

            duration = dur.find('span', class_='duration').text.strip()
            price = pri.text.replace(',','.')
            price = price.replace('zł', '')
            price = price.replace('+', '')
            price = price.split()[-1]

            tmp.append((name, duration, loc, price))
```

```python
import pandas as pd

col_names = ['Description', 'Duration', 'Location', 'Price']

df = pd.DataFrame(tmp, columns=col_names)
```
